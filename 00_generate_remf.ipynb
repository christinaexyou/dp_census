```python
import numpy as np
import pandas as pd
import import_ipynb
import reconstruction_model
```

```python
# Reading in sf1 .csv files into dfs and storing them in a dictionary 
path = '/Users/christinaxu/Documents/dp_census/'
sf1_tables = {}

sf1_table_list = ['P1', 'P8', 'P9', 'P10', 'P11', 'P12', 'P12A', 'P12B', 'P12C', 'P12D', 'P12E', 'P12F', 'P12G', 'P12H', 'P12I', 'P14']

for table in sf1_table_list:
    df=pd.read_csv(path + '/data/mass_sf1/table_{}.csv'.format(table))
    sf1_tables[table] = df
```

```python
# Reading in sf1 .csv files into dfs and storing them in a dictionary 
dhc_tables = {}

dhc_table_list = ['P1', 'P8', 'P9', 'P10', 'P11', 'P12', 'P12A', 'P12B', 'P12C', 'P12D', 'P12E', 'P12F', 'P12G', 'P12H', 'P12I']

for table in dhc_table_list:
    df=pd.read_csv(path + '/data/mass_dhc/table_{}.csv'.format(table))
    dhc_tables[table] = df
```

```python
#t = dhc_tables['P1'].set_index(['COUNTY', 'TRACT', 'BLOCK']).P10001.sort_values(ascending=False )
# rows = t[t>0].index
rows = pd.DataFrame(index=range(5,10)).index
t = sf1_tables['P1'].set_index(
    ['BLOCK', 'COUNTY', 'TRACT']).P0010001.sort_values(
    ascending=False)
rows = rows.append(t[t>0].index).drop_duplicates()
```

```python
results_sf1 = {}

i_split = 1000
n_splits = 100

for i, location_tuple in enumerate(rows[i_split::n_splits]):
    print(i, location_tuple)
    if i % 60 == 0:
        print('.', end=' ', flush=True)
    results_sf1[location_tuple] = reconstruction_model.reconstruct_block(sf1_tables, *location_tuple)

pd.concat(results_sf1).to_csv(f'{path}/sf1_{n_splits}_{i_split}.csv')
```
